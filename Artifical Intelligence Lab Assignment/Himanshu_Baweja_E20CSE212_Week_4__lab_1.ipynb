```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
data = pd.read_csv("dataset.csv")
feature_list=['Hour','Immobilized_bus','Broken_Truck','Vehicle_excess','Accident_victim','Running_over','Fire_vehicles','Occurrence_involving_freight','Incident_involving_dangerous_freight','Lack_of_electricity','Fire','Point_of_flooding','Manifestations','Defect_in_the_network_of_trolleybuses','Tree_on_the_road','Semaphore_off','Intermittent_Semaphore','Slowness_in_traffic_percent']
data.columns = feature_list
data
```

```python
morning =['07:30' ,'08:00', '08:30', '09:00', '09:30','10:00' ,'10:30' ,'11:00','07:00']
noon = ['11:30','12:00' ,'12:30' ,'13:00']
afternoon =['13:30', '14:00', '14:30' ,'15:00' ,'15:30' ,'16:00']
evening =['16:30', '17:00' ,'17:30' ,'18:00']
night =['18:30' ,'19:00' ,'19:30' ,'20:00']
data['Hour'].replace(to_replace = morning,value ="Morning",inplace =True)
data['Hour'].replace(to_replace = noon,value ="Noon",inplace =True)
data['Hour'].replace(to_replace = afternoon,value ="Afternoon",inplace =True)
data['Hour'].replace(to_replace = evening,value ="Evening",inplace =True)
data['Hour'].replace(to_replace = night,value ="Night",inplace =True)
data = pd.get_dummies(data, columns = ['Hour'])
data
```

```python

X = data[['Immobilized_bus', 'Broken_Truck', 'Vehicle_excess', 'Accident_victim',
       'Running_over', 'Fire_vehicles', 'Occurrence_involving_freight',
       'Incident_involving_dangerous_freight', 'Lack_of_electricity', 'Fire',
       'Point_of_flooding', 'Manifestations',
       'Defect_in_the_network_of_trolleybuses', 'Tree_on_the_road',
       'Semaphore_off', 'Intermittent_Semaphore', 'Hour_Afternoon', 'Hour_Evening',
       'Hour_Morning', 'Hour_Night', 'Hour_Noon']]
X
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
Y = data.Slowness_in_traffic_percent
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X[:,17:22], Y, train_size=108, shuffle=False)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print(f"\nMSE: {mse}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X[:,17:22])
x_train1, x_test1, y_train1, y_test1 = train_test_split(X_poly, Y, train_size=108, shuffle = False)
model = LinearRegression()
model.fit(x_train1,y_train1)
y_pred = model.predict(x_test1)
mse = mean_squared_error(y_test1,y_pred)
print(f"\nMSE: {mse}")
rmse = mean_squared_error(y_test1,y_pred,squared=False)
print(f"\nRMSE: {rmse}")
mea = mean_absolute_error(y_test1,y_pred)
print(f"\nMEA: {mea}")
mdea = median_absolute_error(y_test1,y_pred)
print(f"\nMDEA: {mdea}")
r2 = r2_score(y_test1,y_pred)
print(f"\nR2 Square: {r2}")
```

```python
percentage = [0.1,0.2,0.3,0.4]
for i in percentage:
    poly = PolynomialFeatures(degree = 2)
    X_poly = poly.fit_transform(X[:,17:22])
    x_train1, x_test1, y_train1, y_test1 = train_test_split(X_poly, Y, test_size=i, shuffle=False)
    model = LinearRegression()
    model.fit(x_train1,y_train1)
    y_pred = model.predict(x_test1)
    mse = mean_squared_error(y_test1,y_pred)
    print("MSE for %i percent test size is: %f "%(i*100,mse))
```

```python
dg = [2,3,4,5,6,7,8,9,10,15,20]
percentage = [0.5,0.6,0.7,0.8,0.9]
for i in dg:
    poly = PolynomialFeatures(degree = i)
    X_poly = poly.fit_transform(X[:,17:22])
    for j in percentage:
        x_train1, x_test1, y_train1, y_test1 = train_test_split(X_poly, Y, train_size=j, shuffle=False)
        model = LinearRegression()
        model.fit(x_train1,y_train1)
        y_pred = model.predict(x_test1)
        mse = mean_squared_error(y_test1,y_pred)
        print("MSE for polynomial degree %i and training data size %i percent is: %f "%(i,j*100,mse))
    print("\n")
```
