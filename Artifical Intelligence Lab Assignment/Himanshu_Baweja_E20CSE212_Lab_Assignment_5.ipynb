```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import minmax_scale
```

```python
data = pd.read_csv("student-mat.csv",sep=";")
data
```

```python
len(data.columns)
```

```python
new_data =data.drop(['school','sex','Pstatus','Medu','Fedu','Mjob','Fjob','guardian','traveltime','nursery','famrel','Dalc','Walc'],axis = 1)
```

```python
new_data
```

```python
len(new_data.columns)
```

```python
X = pd.get_dummies(new_data, columns=["address", "famsize", "reason", "schoolsup", "famsup", "paid", "activities", "higher", "internet", "romantic"], drop_first=True)
X
```

```python
Y=data.iloc[:,-1]
Y
```

```python
X2 = X.columns
for i in X2:
    X[i] = minmax_scale(X[i])
X
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
y_pred = model.predict(X)
```

```python
# Print mean squared error
mse = mean_squared_error(Y,y_pred)
print(f"\nMSE: {mse}")
```
