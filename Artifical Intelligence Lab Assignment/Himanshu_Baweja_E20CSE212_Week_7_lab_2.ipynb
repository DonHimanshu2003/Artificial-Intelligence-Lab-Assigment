```python
from sklearn.preprocessing import LabelEncoder
df['quality']= LabelEncoder().fit_transform(df['quality'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
mms.fit_transform(df)
```

```python
df=pd.DataFrame(df)
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1:]
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)
ypred=clf.predict(x_test)
```

```python
from sklearn.metrics import precision_score,f1_score,recall_score,roc_auc_score,accuracy_score
p=precision_score(y_test,ypred,average=None)
f=f1_score(y_test,ypred,average=None)
r=recall_score(y_test,ypred,average=None)
roc=roc_auc_score(Y,clf.predict_proba(X),multi_class="ovr")
```

```python
print("Precision:",p)
print("F1 SCORE:",f)
print("RECALL SCORE:",r)
print("ROC:",roc)
```
